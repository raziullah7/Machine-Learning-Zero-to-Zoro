Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

Load the dataset

```python
dataset = pd.read_csv('titanic.csv')
xMatrix = dataset.iloc[:, :]
```

Identify the categorical data

```python
columnsWithCategoricalData = ['Sex', 'Embarked', 'Pclass']
```

Implement an instance of the ColumnTransformer class

```python
columnTransformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), columnsWithCategoricalData)], remainder = 'passthrough')
```

Apply the fit_transform method on the instance of ColumnTransformer

```python
xMatrixCategorized = columnTransformer.fit_transform(xMatrix)
```

Convert the output into a NumPy array

```python
X = np.array(xMatrixCategorized)
```

Use LabelEncoder to encode binary categorical data

```python
yVctr = dataset.iloc[:,1]
lEncoder = LabelEncoder()
y = lEncoder.fit_transform(yVctr)
```

Print the updated matrix of features and the dependent variable vector

```python
print(X)
print(y)
```
